````python
#Calculer la moyenne du prix
mean_price = df_merge_final['price'].mean()
print(f"Moyenne du prix: {mean_price:.2f}")

#Calculer l'écart-type du prix
std_dev_price = df_merge_final['price'].std()
print(f"Écart-type du prix: {std_dev_price:.2f}")

#Calculer le Z-score
df_merge_final['z_score_price'] = (df_merge_final['price'] - mean_price) / std_dev_price
print("Z-scores calculés et ajoutés à la colonne 'z_score_price'.")
````

````python
#Quel est le seuil prix dont le z-score est supérieur à 3?
price_threshold_z_score_3 = df_merge_final[df_merge_final['z_score_price'] > 3]['price'].min()
print(f"Le seuil de prix dont le Z-score est supérieur à 3 est : {price_threshold_z_score_3:.2f}")
````

````python
#Définir un seuil pour les articles "outliers" en prix
Q1 = df_merge_final['price'].quantile(0.25)
Q3 = df_merge_final['price'].quantile(0.75)
IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

print(f"Q1 (25e percentile): {Q1:.2f}")
print(f"Q3 (75e percentile): {Q3:.2f}")
print(f"IQR (InterQuartile Range): {IQR:.2f}")
print(f"Seuil supérieur pour les outliers: {upper_bound:.2f}")
print(f"Seuil inférieur pour les outliers: {lower_bound:.2f}")
````

````python
#Définir le nombre d'articles et la proportion de l'ensemble du catalogue "outliers"
outliers = df_merge_final[(df_merge_final['price'] < lower_bound) | (df_merge_final['price'] > upper_bound)]
num_outliers = outliers.shape[0]
total_articles = df_merge_final.shape[0]
proportion_outliers = (num_outliers / total_articles) * 100

print(f"Nombre d'articles outliers en prix : {num_outliers}")
print(f"Proportion d'articles outliers dans le catalogue : {proportion_outliers:.2f}%")
````

Cette relation est modérée, ce qui indique que le prix n'est pas le seul facteur déterminant les ventes, mais qu'il joue un rôle important.

*   Les produits les moins chers ont tendance à générer plus de ventes, ce qui est logique. Cependant, la corrélation n'est pas extrêmement forte, suggérant que des produits plus chers peuvent aussi bien se vendre en fonction de leur valeur perçue ou de leur spécificité.
*   La gestion des stocks semble assez réactive aux ventes, mais pas directement liée au prix. Une stratégie pourrait être de s'assurer que les produits à forte demande (qui se vendent bien) sont toujours suffisamment en stock, sans que cela ne dépende de leur prix.
*   Il n'y a pas de signal clair montrant que des stocks élevés ou faibles influencent directement le prix, ce qui pourrait être une opportunité. Par exemple, si un produit a un stock très élevé et se vend lentement, une baisse de prix pourrait stimuler les ventes (compte tenu de la corrélation négative entre prix et ventes), mais cette décision ne serait pas dictée par la seule quantité de stock.
